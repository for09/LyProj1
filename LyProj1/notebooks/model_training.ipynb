```python
from sklearn.model_selection import TimeSeriesSplit
def split_df_KNN(X,y):
 

    tscv = TimeSeriesSplit(n_splits = 6)


    X_train_list=[]
    X_test_list=[]
    y_train_list=[]
    y_test_list =[]

    for train_index , test_index in tscv.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        y_train = y_train.values.ravel()
        y_test = y_test.values.ravel()

        X_train_list.append(X_train)
        X_test_list.append(X_test)
        y_train_list.append(y_train)
        y_test_list.append(y_test)

        # print(X_train_list)
        # print(X_test_list)
        # print(y_train_list)
        # print(y_test_list)
    return X_train_list,X_test_list,y_train_list,y_test_list
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score,roc_curve,auc
import matplotlib.pyplot as plt
import seaborn as sns

def knn_str(df, company_name, indicator):
    company_data = df[df['company_name'] == company_name].copy()
    X = company_data[[indicator]]
    y = company_data[['signal_' + indicator]]
    X_train_list, X_test_list, y_train_list, y_test_list = split_df_KNN(X, y)
        
    knn = KNeighborsClassifier(n_neighbors = 7 )
    
    all_prediction = []
    avg_precision = []
    avg_recall = []
    avg_F1 = []
    # fpr_list = []
    # tpr_list = []
    # roc_auc_list = []
  
    for X_train, X_test, y_train, y_test in zip(X_train_list, X_test_list, y_train_list, y_test_list):
        knn.fit(X_train, y_train)
        
        prediction = knn.predict(X_test)
        threshhold = 0.5
        binary_prediction  = np.where(prediction > threshhold, 1, 0)
        
        all_prediction.extend(binary_prediction)
        
        precision = precision_score(y_test, binary_prediction, average='micro')
        
        recall = recall_score(y_test, binary_prediction, average='micro')
        
        f1 = f1_score(y_test, binary_prediction, average='micro')
        
        avg_precision.append(precision)
        avg_recall.append(recall)
        avg_F1.append(f1)
    
        # fpr, tpr, _ = roc_curve(y_test, prediction)
        # roc_auc = auc(fpr, tpr)
        # fpr_list.append(fpr)
        # tpr_list.append(tpr)
        # roc_auc_list.append(roc_auc)


    # plt.figure(figsize=(4, 3))
    # for i in range(len(fpr_list)):
    #     plt.plot(fpr_list[i], tpr_list[i], label='ROC curve (area = %0.2f)' % roc_auc_list[i])
    # plt.plot([0, 1], [0, 1], 'k--')
    # plt.xlim([0.0, 1.0])
    # plt.ylim([0.0, 1.05])
    # plt.xlabel('False Positive Rate')
    # plt.ylabel('True Positive Rate')
    # plt.title('Receiver Operating Characteristic (ROC) Curve')
    # plt.legend(loc="lower right")
    # plt.show()

    avgr_precision = np.mean(avg_precision)
    avgr_recall = np.mean(avg_recall)
    avgr_f1 = np.mean(avg_F1)
    final_pred = 1 if np.mean(binary_prediction) >= 0.5 else 0
    
    return final_pred, avgr_precision, avgr_recall, avgr_f1

# Example usage:
company_name = 'AMAZON'
indicator = 'SMA'
prediction, precision, recall, f1 = knn_str(df, company_name=company_name, indicator=indicator)
print('Recommendation for', company_name, 'based on', indicator, ':', 'buy' if prediction == 1 else 'sell')
print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1)
```
